## 量化与价投之辨

## 我的基本情况
  炒股经验2年，累计收益率50%，在连续两年大行情下其实不行，基本是在学东西
  
  都学习并相信价值投资理论，对

## 传统价值投资的一些特点：

 - 低换手率
 - 长期持股
 - 选股>择时
   极端的人都不择时
 - 对买入的股票都有深入研究
 - 短期走势难以预测
 - 把主要精力放在公司研究
 - 试图利用市场之外的信息盈利
 - 会研究企业运营、社会发展、公司文化、管理层人品等信息
 - 市场的表现几乎不会影响操作选择，会考虑市场其他投资者的行为逻辑
 - 资金量天花板很高

## 常见量化的一些特点：
 - 高换手率，高频交易更甚
 - 定期轮换
 - 试图单从历史交易数据预测未来
 - 试图提高择时的影响
 - 人不需要了解买入的公司
 - 对公司的研究止于基础财务数据、市场数据等，几乎没有市场之外的信息
 - 对公司运营几乎一无所知
 - 将市场表现作为一个客观事件，认为自己的交易行为不会对市场表现造成影响
 - 资金量天花板较低

## 目前我个人的一些理解
 - 量化交易已经发展到没有人可以质疑了，但是传统价值投资与量化交易在各自领域深入时差异加大，而不是殊途同归
 - 量化交易真正超额收益的部分还是对市场无效的利用来贡献的
 
 在市场有效理论中，如果可以单独依赖历史信息（交易信息）就取得超额收益，那市场无效。从这个角度上讲，量化交易的基础就是市场无效，价值投资是认为市场弱势有效或者半强有效。市场十分复杂，像我们人类社会一样复杂，所以用简单3个等级描述其有效性也不科学，市场很可能在一部分时段和一部分股票上在特定情况下一定程度上有效，其他时候无效。
 - 对计算机技术的使用，对技术的应用也不是量化的专利
 
 很多量化其实还是在依照价值投资的理论，比如财务数据综合评分，应该是利用IT技术进行价值投资，不过在技术上还是离量化研究近一点，这可能也是2者的交集吧。
 - 底层逻辑
  
  价值投资的东西的底层逻辑是很顺的。量化就不一样了，虽然我们开发出这么多评价量化策略的方法，但是还是很难避免随机性的影响
 - 投资与生活的关系
 
 传统价投更容易利用日常时间，生活中选股这一思路使得投资生活与日常生活不远，没有不合群的风险；而量化的感觉和做技术面是一样的，看起来像神棍，从这个角度看，量化交易牺牲更多，必须要取得比价投更高的回报才是划算的
 - 收益的根本来源
 
 价值投资更多是赚取合理范围内的钱，在收益和风险的选择中更保守，超额收益的部分承认是收割市场情绪，理性收割不理性；而量化方法本身就要求尽可能去追求最高的/最后一部分利润，奇淫巧技所以对风险更开放
 - 投资者与市场的关系
 
 股市是由大量交易者参与构成的，价值投资是把自身看作趋势构造的参与者的，而量化投资似乎把自身看作局外人，是从趋势中揩油的，毕竟揩油不会影响原来的走势，如果影响了原来的走势，那还怎么用
 - 对短线信息的认识
 
 价投认为短线，1个月以内的波动都是垃圾信息，“市场大部分时候都在做随机运动”，但在量化里面1个月很多策略都可以调仓好几次了
 - 手续费/摩擦成本的影响
 
 延后缴税是巴菲特投资理论重要的一点，a股也有鼓励长持的红利税政策，但这些在量化中都是不重要的
 - 天花板
 
 - 对买入标的的理解
 
 在了解公司这个层面上，量化是一点都没有的，把持股看作当公司的股东，也是没有的；可能会导致持股不坚定的问题，就是在高波动时，理解不透导致无法持股；缺点也体现在无法充分集中，就算集中了也很难解释超额收益的来源
 
 - 基础假设方面
 
 格雷厄姆、巴菲特基本假设“赚尽市场的每一分钱是不可能的”，虽然没有具体论证，但是在他们的著作中基本都在表示这样的想法，例如不要想着一手长线一手短线，牛市逃顶熊市抄底，“盈亏同源”深层次的理解就是这样的。但是量化不是，我想每一个做量化的人大概潜意识里都认为可能存在一个完美策略，这个策略可以在任何回测周期跑赢基准，如果没有这个追求的话，那量化在做什么呢？所以这也是一个核心分歧，一个鼓励守拙，另一个要奋进。
 

## 策略观察上的发现
 - 长期持有在量化策略也有效
 
    长期主义是价投的基本之一，所以在量化上，有一些需要将持股周期作为一项选择的策略，如果拉长持股周期，会出现明显的业绩提升；这也不知道是价投的胜利，还是量化的胜利
 - 真正有效策略的时效很短
 
    量化基金的运行应该是需要不断根据市场情况进行更新的，模型更新，算法更新，使用方式更新等等；可能需要很多市场上的经验来做，基于其他理论判断市场情况来选择合适的模式，这时候具体发挥作用的时候到底是哪个逻辑
 - 广泛存在的隐形未来函数

   辨别隐性未来函数可以写一个大专题了
   
   例如：
   - 根据某些基金评比的结果进行回测，基金评比本身就会选择表现优秀的基金
   - 回测时间段精准选择
   - 算法模型训练数据与回测数据重叠等等
 - 理想回测场景
 
 到底极端行情是测试量化策略的理想场地，还是平静的行情，很难讲，不管是量化还是价投，总有被时代选中的投资策略
 - 不存在的完美策略
   
   在任何行情、任何股票、任何回测周期内都能稳定赚钱，收益高、回撤小、alpha高beta小，信息比率高，波动小
 - 对策略的评价
 
 评价基金经理时的道德风险对量化策略依然适用，即量化策略天生就被要求高换手率，一定要持仓和操作逻辑让人理解不了才能显示其高端；而这种隐形要求可能刚好使其无法表现
 - 如何看待股市
 
 股市的出现和运行是有其底层逻辑的，企业融资便利、交易形成价格等等，如果做量化完全忽略这些底层逻辑，只利用最表象的特征，很难让人信服

## 技术指标的量化

  技术指标的底层仍然是心理学，量化后的仍然是心理学，其他都是表象

## 需要不断进行研究的

- 策略评价引入更多指标：年均换手率、最长持股周期、持有过的股票、收益贡献股票分布
- 对一些重要因子在其他条件下的回测：拉长换手周期、拉长评价周期
- 某些单因子的表现存在明显两头收益高中间收益低的问题，例如pe、股息率等，值得深入研究

- 各种策略可以作为检验流行理论的一种方式
- 真正赚钱的策略可能是需要及时调整的，感受行情并及时调整策略仍然需要经验or研究，这跟机器学习在其他领域的应用也类似，所以要两条腿走路
- 矛盾并不在价投或者量化，而在于研究的深入程度
